> *Do. Not. Ever. Install pyenv using apt.*
>
> That would assign it root privileges that will break its expected behaviour as a userspace environment configurator
>
>  You must install it using the method explained [HERE](https://github.com/pyenv/pyenv/?tab=readme-ov-file#linuxunix)
>
