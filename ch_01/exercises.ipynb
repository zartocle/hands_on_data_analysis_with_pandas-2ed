## Exercise 5: Calculating statistics and verifying
### mean

```python
def calculate_mean_value(my_list:list):
    return sum(my_list)/len(my_list)

calculate_mean_value(salaries)
```

```python
st.mean(salaries) == calculate_mean_value(salaries)
```

### mode

```python
from collections import Counter
c = Counter()
```

```python
for i in salaries:
    c[i]+=1
```

```python
c.most_common(1)
```

```python
# Alternative, more syntethic way of using the Counter object:
Counter(salaries).most_common()[0][0] == st.mode(salaries)
```

### sample variance
Remember to use Bessel's correction.

```python
import math

def variance_calculator(items_list_par:list):
    var_numerator = 0
    abs_value_mean = abs(calculate_mean_value(items_list_par))
    for i in items_list_par:
        squared_diff = (i - abs_value_mean) ** 2
        var_numerator = var_numerator + squared_diff

    return var_numerator / (len(items_list_par) - 1 )

varianzanza = variance_calculator(salaries)
```

```python
varianzanza == st.variance(salaries)
```

### sample standard deviation
Remember to use Bessel's correction.

```python
math.sqrt(varianzanza)
```

```python
math.sqrt(varianzanza) == st.stdev(salaries)
```
